```python
import h5py
import pandas as pd
```

```python
file_name = 'N-CMAPSS_DS02-006.h5'
file_path = '/content/Data/data_set/data_set/data_set/'+ file_name
```

```python
h5py_keys =  []

with h5py.File(file_path, 'r') as file:
    for key in file.keys():
        h5py_keys.append(key)

```

```python
(h5py_keys)
```

```python
keys_dict = {'A_dev':'A_var',
             'A_test':'A_var',
             'T_test': 'T_var',
             'T_dev': 'T_var',
             'W_dev': 'W_var',
             'W_test': 'W_var',
             'X_s_dev':'X_s_var',
             'X_s_test':'X_s_var',
             'X_v_dev':'X_v_var',
             'X_v_test':'X_v_var',
             }
```

```python
import pandas as pd
import h5py
import numpy as np
import gc

dev_dataframes = []
test_dataframes = []

with h5py.File(file_path, 'r') as file:
    for key in h5py_keys:
        if key not in keys_dict.values():
            data = file[key][:]
            if key != 'Y_dev' and key != 'Y_test' and key in keys_dict:
                _columns = [x.decode() for x in file[keys_dict[key]][:]]
            elif key == 'Y_dev' or key == 'Y_test':
                _columns = ['Y']
            else:
                continue

            df = pd.DataFrame(data, columns=_columns)
            if('dev' in key):
              dev_dataframes.append(df)
            elif('test' in key):
              test_dataframes.append(df)

            del data
            gc.collect()

if dev_dataframes:
    df_dev = pd.concat(dev_dataframes, axis=1)

if test_dataframes:
    df_test = pd.concat(test_dataframes, axis=1)


del dev_dataframes
del test_dataframes
gc.collect()
```

```python
df_dev['split'] = ['dev']* len(df_dev)
df_test['split'] = ['test']* len(df_test)

df = pd.concat([df_dev, df_test], axis = 0)

del df_dev
del df_test
gc.collect()
```
