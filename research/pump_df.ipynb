```python
import pandas as pd
import re
```

```python
with open('/content/drive/MyDrive/Projects/LLMs/pump.txt', 'r') as file:
    text = file.read()
```

```python
column_pattern = r"repair_logs\s\((.*?)\)\sVALUES"
data_pattern = r"\((.*?)\)"
```

```python
columns_match = re.search(column_pattern, text, re.S)
columns = columns_match.group(1).split(', ') if columns_match else []
```

```python
rows = re.findall(data_pattern, text, re.S)
```

```python
data = []
for row in rows:
    row = re.sub(r"'(.*?)'", lambda m: m.group(1).replace(',', '###'), row)
    values = [v.strip().replace('###', ',') for v in row.split(",")]
    data.append(values)
```

```python
df = pd.DataFrame(data, columns=columns)
```
